```python
import os
import glob
from scipy.io import wavfile
import matplotlib.pyplot as plt
import IPython.display
%matplotlib inline
```

```python
target = '../gen/*.wav'
for f in sorted(glob.glob(os.path.join(target))):
    print(f)
    sr, y = wavfile.read(f)
    plt.figure()
    plt.plot(y)
    plt.show()
    IPython.display.display(IPython.display.Audio(f))
```
